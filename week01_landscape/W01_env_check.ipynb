```python
import sys
import platform
from datetime import datetime

print("Environment Check Timestamp:", datetime.now().isoformat(timespec="seconds"))
print("Python executable:", sys.executable)
print("Python version:", sys.version)
print("Platform:", platform.platform())

major, minor = sys.version_info[:2]

if (major, minor) == (3, 11):
    print("PASS: Python 3.11 is active.")
else:
    raise RuntimeError(f"Expected Python 3.11, but found Python {major}.{minor}. Please switch the notebook kernel.")
```

```python
import importlib
from importlib.metadata import version, PackageNotFoundError
import pandas as pd

packages = [
    {"display": "JupyterLab", "import_name": "jupyterlab", "package_name": "jupyterlab"},
    {"display": "pandas", "import_name": "pandas", "package_name": "pandas"},
    {"display": "NumPy", "import_name": "numpy", "package_name": "numpy"},
    {"display": "scikit-learn", "import_name": "sklearn", "package_name": "scikit-learn"},
    {"display": "PyTorch", "import_name": "torch", "package_name": "torch"},
    {"display": "Gymnasium", "import_name": "gymnasium", "package_name": "gymnasium"},
    {"display": "Stable-Baselines3", "import_name": "stable_baselines3", "package_name": "stable-baselines3"},
    {"display": "PettingZoo", "import_name": "pettingzoo", "package_name": "pettingzoo"},
    {"display": "PyOD", "import_name": "pyod", "package_name": "pyod"},
    {"display": "plotly", "import_name": "plotly", "package_name": "plotly"},
    {"display": "Streamlit", "import_name": "streamlit", "package_name": "streamlit"},
]

results = []

for pkg in packages:
    display = pkg["display"]
    import_name = pkg["import_name"]
    package_name = pkg["package_name"]

    try:
        module = importlib.import_module(import_name)

        try:
            pkg_version = version(package_name)
        except PackageNotFoundError:
            pkg_version = getattr(module, "__version__", "version not found")

        results.append({
            "Package": display,
            "Import Name": import_name,
            "Version": pkg_version,
            "Status": "PASS"
        })

    except Exception as e:
        results.append({
            "Package": display,
            "Import Name": import_name,
            "Version": None,
            "Status": f"FAIL: {type(e).__name__}: {e}"
        })

env_df = pd.DataFrame(results)
env_df
```

```python
failed = env_df[env_df["Status"] != "PASS"]

if failed.empty:
    print("PASS: All required packages imported successfully.")
else:
    print("FAIL: Some required packages did not import successfully.")
    display(failed)
    raise RuntimeError("Environment check failed. Please install or fix the failed packages.")
```

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=42)

sample_df = pd.DataFrame({
    "robot_id": ["aido_rover_001", "aido_rover_002", "aido_rover_003"],
    "battery_soc": rng.uniform(0.35, 0.95, size=3),
    "motor_current": rng.normal(12.0, 1.5, size=3),
    "wifi_rssi": rng.normal(-65.0, 4.0, size=3)
})

sample_df["fleet_health_score"] = (
    0.5 * sample_df["battery_soc"]
    + 0.3 * (1 / (1 + np.exp((sample_df["motor_current"] - 12) / 2)))
    + 0.2 * (1 / (1 + np.exp(-(sample_df["wifi_rssi"] + 70) / 5)))
)

sample_df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest

feature_cols = ["battery_soc", "motor_current", "wifi_rssi", "fleet_health_score"]

X = sample_df[feature_cols]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = IsolationForest(random_state=42, contamination=0.34)
labels = model.fit_predict(X_scaled)

sample_df["sklearn_iforest_label"] = labels

print("PASS: scikit-learn preprocessing and IsolationForest smoke test completed.")
sample_df
```

```python
import torch

tensor = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
result = tensor @ tensor.T

print("PyTorch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())
print("Tensor result:")
print(result)

assert result.shape == (2, 2)

print("PASS: PyTorch tensor smoke test completed.")
```

```python
import gymnasium as gym

env = gym.make("CartPole-v1")
obs, info = env.reset(seed=42)

print("Observation shape:", obs.shape)
print("Initial observation:", obs)

action = env.action_space.sample()
next_obs, reward, terminated, truncated, info = env.step(action)

env.close()

print("Sample action:", action)
print("Reward:", reward)
print("Terminated:", terminated)
print("Truncated:", truncated)
print("PASS: Gymnasium smoke test completed.")
```

```python
from pyod.models.iforest import IForest

pyod_model = IForest(random_state=42, contamination=0.34)
pyod_model.fit(X_scaled)

pyod_scores = pyod_model.decision_scores_
sample_df["pyod_iforest_score"] = pyod_scores

print("PASS: PyOD Isolation Forest smoke test completed.")
sample_df
```
